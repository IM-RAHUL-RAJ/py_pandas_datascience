```
import matplotlib.pyplot as plt
```

```
x=[1,2,3,4,5,6,7]
y=[4,5,7,8,9,0,3]
```

```
plt.plot(x,y,"r+-") #different markser
```

```
plt.plot(x,y,"rD") #different markser
```

```
plt.plot(x,y,color="red",marker="D",linestyle="-",markersize=10) #different markers
```

```
plt.plot(x,y,color="red",marker="D",linestyle="-",markersize=10,alpha=0.4) #different markers
```
