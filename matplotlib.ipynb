```
import matplotlib.pyplot as plt
```

```
x=[1,2,3,4,5,6,7]
y=[4,5,7,8,9,0,3]
```

```
plt.plot(x,y,"r+-") #different markser
```

```
plt.plot(x,y,"rD") #different markser
```

```
plt.plot(x,y,color="red",marker="D",linestyle="-",markersize=10) #different markers
```

```
plt.plot(x,y,color="red",marker="D",linestyle="-",markersize=10,alpha=0.4) #different markers
```

**axes,label,legend grid**

```
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,46,47,49]
min_t=[43,45,40,33,35,36,38]
avg_t=[45,46,48,49,41,42,46]
#weather data
```

```
plt.plot(days,max_t)
```

```
plt.xlabel('Days')
plt.ylabel('temp')
plt.title("sample weather data")

plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")

plt.legend(loc="upper right")
#by default plt.legend(loc="best")
```
